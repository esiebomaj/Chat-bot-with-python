```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
import numpy as np
import random
import json
import pandas as pd
import nltk
import string
import pickle
```

```python
with open('intents - Jerry.json') as file:
    data=json.load(file)
```

```python
try:
    with open('datal.pickle','rb') as f:
        words, labels, training, output = pickle.load(f)
    
    
except:
    words=[]
    labels=[]
    docs_x=[]
    docs_y=[]
    
    stemmer=nltk.lancaster.LancasterStemmer()
    
    for intent in data['intents']:
        for pattern in intent['patterns']:
            wrds=nltk.word_tokenize(pattern)
            wrds=[w for w in wrds if w not in string.punctuation]
            words.extend(wrds)
            docs_x.append(wrds)
            docs_y.append(intent['tag'])

        if intent['tag'] not in labels:
            labels.append(intent['tag'])
            
    words=[stemmer.stem(w.lower()) for w in words]
    words=sorted(list(set(words)))
    labels=sorted(labels)
    training=[]
    output=[]
    out_empty=[0 for i in labels ]
    
    
    for x, doc in enumerate(docs_x):
        bag=[]
        wrds=[stemmer.stem(w) for w in doc if w not in string.punctuation]
        for word in words:
            if word in wrds:
                bag.append(1)
            else:
                bag.append(0)
        output_row=out_empty[:]
        output_row[labels.index(docs_y[x])]=1

        training.append(bag)
        output.append(output_row)
    
    training=np.array(training)
    output=np.array(output)
    
    
with open('data_pickle','wb') as f:
    pickle.dump((words, labels, training, output),f)
```

```python
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(26,input_shape=(56,)),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(9, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['categorical_accuracy'])
```

```python
model.summary()
```

```python
try:
    model=keras.models.load_model('modell.tf')
except:
    model.fit(training,output,epochs=1000,verbose=1)
    model.save('model.tf')
```

```python
def bag_of_words(s, words):
    bag=[0 for i in words]
    s_words=nltk.word_tokenize(s)
    s_words=[stemmer.stem(w.lower()) for w in s_words]
    
    for x,w in enumerate(words):
        if w in s_words:
            bag[x]=1
    return np.array(bag)
```

```python
import random
def chat():
    print('#'*45)
    print('*'*5+'This is jeremiah\'s personal chat-bot'+'*'*5)
    print('start talking with the bot(type quit to exit)')
    while True:
        inp=input('you:')
        if inp.lower()=='quit':
            break
        else:
            results=model.predict([[bag_of_words(inp,words)]])
#            print(bag_of_words(inp,words))
#            print(results)
#            print(results[np.argmax(results)])
                
            tag=labels[np.argmax(results)]
      #            print(labels[np.argmax(results)])

            for i in data['intents']:
                if i['tag']==tag:
                    responses=i['responses']

            print('Jerry\'s Bot : '+random.choice(responses)+'\n')

chat()
```
